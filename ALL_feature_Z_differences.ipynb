```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
data=pd.read_csv('train_data.csv')
labels=pd.read_csv('train_labels.csv')
```

```python
labeled_data=data.set_index('customer_ID').join(labels.set_index('customer_ID'))
```

```python
labeled_data.to_csv("labeled_data.xlsx")
```

```python
l=[]
errors =[]
for i in labeled_data.columns[1:]:
    try:
        print(i,
              labeled_data.loc[labeled_data['target'] == 1, i].mean(),
              labeled_data.loc[labeled_data['target'] == 0, i].mean(),
              labeled_data[i].std())
        l.append(i)
        l.append(labeled_data.loc[labeled_data['target'] == 1, i].mean())
        l.append(labeled_data.loc[labeled_data['target'] == 0, i].mean())
        l.append(labeled_data[i].std())
    except:
        errors.append(i)
        print("error",i)
```

```python
feature=[]
target_mean=[]
non_target_mean=[]
std_dev=[]
for i in range(len(l)):
    if i%4==0:
        feature.append(l[i])
    elif i%4==1:
        target_mean.append(l[i])
    elif i%4==2:
        non_target_mean.append(l[i])
    elif i%4==3:
        std_dev.append(l[i])
    else:
        print("error",i,l[i])
        
    
```

```python
newdf=pd.DataFrame()
```

```python
newdf["feature"]=feature
newdf["target_mean"]=target_mean
newdf["non_target_mean"]=non_target_mean
newdf["std_dev"]=std_dev
newdf.head()
```

```python
newdf.to_csv("Feature_DefMean_NondefMean_std.csv")
```

```python
Z_diff=[]
for i in range(len(newdf)):
    x= (newdf.iloc[i,1]-newdf.iloc[i,2])/newdf.iloc[i,3]
    x= math.sqrt(x**2)
    Z_diff.append(x)
```

```python
newdf["Z_diff"]=Z_diff
```
